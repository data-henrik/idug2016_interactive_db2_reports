# How to interface DB2 with simple SQL using "Magic"
***
*  Notebooks have so-called magics
*  There is one magic for SQL statements
*  And it supports DB2!


```python
# loads the SQL magic extensions
%load_ext sql
```

```python
# Connect to dashDB / DB2
%sql db2+ibm_db://user:password@URLtoDB2orDashDB:50000/BLUDB
```

```python
# Simple SQL statement to query DB2 catalog
%%sql 
select tabname,tabschema,colnames from syscat.indexes where uniquerule='P'
```

```python
# Query DB2 and use result to create pie chart
result= %sql select tabname,count(colname) from syscat.columns group by tabname order by 2 desc limit 20
%matplotlib inline
result.pie()
```
